```python
DG = prescription[['DrugName', 'GenericName']]
DG
```

```python
# Assuming you have a prescription name dataframe called "DG"
# with columns: DrugName, GenericName, and other columns

# Convert drug names to lowercase
DG['DrugName'] = DG['DrugName'].str.lower()
DG['GenericName'] = DG['GenericName'].str.lower()

# Trim the part enclosed in square brackets at the front
DG['DrugName'] = DG['DrugName'].str.replace(r'^\[[^\]]+\]\s*', '', regex=True)
DG['GenericName'] = DG['GenericName'].str.replace(r'^\[[^\]]+\]\s*', '', regex=True)
```

```python
# Concatenate drugs with the same name within the "DrugName" column
concatenated_drugs = DG.groupby('DrugName')['DrugName'].transform(lambda x: '\n'.join(x))

# Create a new dataframe with the concatenated drugs for the "DrugName" column
new_df = pd.DataFrame({'DrugName': concatenated_drugs})

# Concatenate drugs with the same name within the "GenericName" column
concatenated_generic = DG.groupby('GenericName')['GenericName'].transform(lambda x: '\n'.join(x))

# Add the concatenated drugs for the "GenericName" column to the new dataframe
new_df['GenericName'] = concatenated_generic

# Count the occurrences of each value in the "DrugName" column
drugname_counts = new_df['DrugName'].value_counts().reset_index()
drugname_counts.columns = ['DrugName', 'DrugName_Count']

# Count the occurrences of each value in the "GenericName" column
genericname_counts = new_df['GenericName'].value_counts().reset_index()
genericname_counts.columns = ['GenericName', 'GenericName_Count']

# Combine the counts into a single dataframe
combined_df = pd.concat([drugname_counts, genericname_counts], axis=1)
```

```python
import re

# Concatenate drugs with the same name within the "DrugName" column
concatenated_drugs = DG.groupby('DrugName')['DrugName'].transform(lambda x: ' '.join(set(x)))

# Create a new dataframe with the concatenated drugs for the "DrugName" column
new_df = pd.DataFrame({'DrugName': concatenated_drugs})

# Concatenate drugs with the same name within the "GenericName" column
concatenated_generic = DG.groupby('GenericName')['GenericName'].transform(lambda x: ' '.join(set(x)))

# Add the concatenated drugs for the "GenericName" column to the new dataframe
new_df['GenericName'] = concatenated_generic

# Remove doses from DrugName
new_df['DrugName'] = new_df['DrugName'].astype(str).apply(lambda x: re.sub(r'\s*\d+(\.\d+)?\s*\w+', '', x))

# Remove doses from GenericName
new_df['GenericName'] = new_df['GenericName'].astype(str).apply(lambda x: re.sub(r'\s*\d+(\.\d+)?\s*\w+', '', x))

# Count the occurrences of each value in the "DrugName" column
drugname_counts = new_df['DrugName'].value_counts().reset_index()
drugname_counts.columns = ['DrugName', 'DrugName_Count']

# Count the occurrences of each value in the "GenericName" column
genericname_counts = new_df['GenericName'].value_counts().reset_index()
genericname_counts.columns = ['GenericName', 'GenericName_Count']

# Combine the counts into a single dataframe
combin_df = pd.concat([drugname_counts, genericname_counts], axis=1)
combin_df
```
